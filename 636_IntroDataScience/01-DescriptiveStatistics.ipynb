## Descriptive Statistics

### Data Sets

- For example, a set of $N$ scalar measurements is notated as $ \displaystyle \big\{x_i\big\}_{i=1}^N $
- We can characterize data on Location, Dispersion, and Shape

```python
# Useful modules we'll always need
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate 1000 normally distributed random data points and print the first element
N = 1000
x = np.random.randn(N)
print (x[0], "\n")
print ("%f, %f, ..., %f" % (x[0], x[1], x[N-1]), "\n") # creating index
print ("%f = %f" % (x[N-1], x[-1]), "\n") # two different ways of getting the last element
print("rows:", x.shape, "columns:", x.size)
```

```python
# Try/Except Blocks for debugging
for i in range(N-3,N+5):
    try: 
        print ("%d : \t %r" % (i, x[i]))
    except IndexError as err: 
        print (err)
```

```python
# plot as a function of the index
plt.plot(x,'bx')
```

```python
# Histogram of points
h = plt.hist(x, 50)
```

### Location

- Mode: greatest count of data (can be unimodal or multimodal)
- Mean $\bar{x}$: $\displaystyle \bar{x} = \frac{1}{N} \sum_{i=1}^N x_i$
    - but indexing starts with 0 in Python and most computer languages: $\displaystyle \bar{x} = \frac{1}{N} \sum_{i=0}^{N-1} x_i$
- Median: The index "middle" of sorted data

```python
mean = np.mean(x)
med = np.median(x)
print("The mean of the data is:",mean,"and the median is:",med)
```

```python
fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot(1,1,1)
ax.hist(x,20)
ax.arrow(mean,0,0,5,color='r')
ax.arrow(med,5,0,5,color='y')
ax.set_xlim(-10, 10);
```

### Dispersion
- Sample variance: $\displaystyle s^2 = \frac{1}{N\!-\!1} \sum_{i=1}^N \left(x_i - \bar{x}\right)^2$
- Standard deviation: $\displaystyle s = \sqrt{s^2}$
    - $N-1$ used for degrees of freedom, an unbiased estimator excludes the parameter

```python
var=np.var(x, ddof=0) #use ddof=0 for population, 1 for sample
std=np.std(x, ddof=0)
print("The variance is:", var, "with St. Dev:", std)
```

### Outliers
- If just one element is too large, e.g., erroneously becomes $+\infty$
- This causes both mean and variance to explode

```python
x[0] = 1e5
plt.hist(x, 100);
```

### Robustness
- Median > Mean
- Median Absolute Deviation (MAD) > Standard Deviation

```python
print ('Average old vs new: %f %f' % (mean, np.mean(x)))
print ('Median  old vs new: %f %f' % (med, np.median(x)))
```
