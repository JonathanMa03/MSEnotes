#### Strings

```python
# for strings, single quotes and double quotes are equivalent
x='dog' 
print(type(x))
print(x) # prints without quotes
x # prints with quotes
```

```python
# Two ways to put quotes in quotes
x1='put something in "quotes" in a string' # direct quoting
print(x1)
x2="string plus \"another\" set of characters" # backslash method
print(x2)
```

```python
# we can see how long a string is, or any object with the len() function
# if a function can be used for different object, its called overloaded
st="dogs and cats"
len(st)
```

```python
# We can refer to particular elements in a string using square brackets with values 0,1,...,length of string-1
st="do you prefer to have a dog or a cat as a pet?"
print(len(st))
print(st[0])
print(st[1])
print(st[-4]) #negative assignment counts from the right
```

```python
# A colon is used to "slice" a string
print(st)
print(st[0:6]) # 0,1,2,3,4
print(st[2:])  # 2,3,... end of string
print(st[:3])  # 0,1,2
print(st[-4:])
```

```python
# the + operator is used to join strings together, and is overloaded as well
x1='cat'
x2='69'
x1+=x2 # shortcut for x1 = x1 + x2
print(x1)
```

```python
# we can force conversions
x=1.1
print(type(x))
x1=str(x) #force float to string
print(type(x1))
```

```python
# we can use the .replace() function to replace string postions
st = 'my dog ate my homework'
s1=st.replace("dog","cat")
print(st) #note the value isn't reassigned, because strings are INMUTABLE
print(s1) #we need assignment to change print values
```

```python
# backslahses are used for special characters
# \n = new line character
# \t = tab character
# \\ = backslash character
# \" = to put quotes inside a string (the backslash "escapes" the meaning of the ")
# \' = to put an apostrophe inside a string
print("function is defined as: \n \t x exists as a \"string.\" so let\'s code!")
```

```python
# useful string functions
str='cAt'
print(str.upper()) # uppercase
print(str.lower()) # lowercase
st='c.a.t'
print(st.split(".")) #removes periods and splits to individual strings
"_".join(["my","file","name"]) # join the three together with underscore
#"my_file_name".split("_") #['my', 'file', 'name']
```

#### Lists

```python
a=[1,2,3,4,5,6] #use brackets for lists, indices work the same way
print(a[0])
print(a[-1])
print(a[0:2]) # up to and not including '2'
print(a[-1:-3]) #slicing only works left to right, so gives nothing
```

```python
# lists can have lists, and indexing works the same way
b=[1,'cat',[3,4,5]]
print(b)
print(b[2])
print(b[2][1])
```

```python
# lists are mutable, and so are elements of the lists
numbers = [1, 2, 3, 4]
print(numbers)

numbers[2] = 99 # Modify an element
print(numbers) # this works, if it was immutable, we would need numbers=numbers[2]
```

```python
# Helpful list methods
a=[6,7,8]
a.insert(1,2) # insert(position, element) and shift everything
print(a)
a.append(5) # add 5 to the end
print(a)
a+=[9,10] #concatenation
print(a)
a.extend([43,52]) # add to end
print(a)
a.remove(52) # remove an element, removing can't take lists
print(a)
a.sort()
print(a) # sorts the list from small to large, or a-z
x=a.pop() # remove last element
print(x, a)
del(a[6]) # delete an element by index
print(a)
```

```python
# Making a list of parameters for a function
def f(x,y,z):
    return(x*y+y*z)
f(1,2,3)
```

```python
# we can wrap the function like this
def fnew(L):
    x=L[0]
    y=L[1]
    z=L[2]
    return(f(x,y,z))
fnew([1,2,3])
```

```python
#Alternative with decorators
L=[1,2,3]
fnew(L)
```

#### Tuples

```python
# Tuples are like lists, but they are immutable and can store anything
L=[1,2,3]
car={"Bob":"Mustang","Judy":"Beetle"}
s="mystring"
f=98.6
b1=True
b2=False

t=(L,s,f,b1,b2,car)
print(t)
print(type(t))
# t[1]=87.2 will give an error. REMEMBER: immutability=no direct assignment
```

```python
# assignment works similarily
t=(2,3,4)
a,b,c=t
print(a,b,c)
```

```python
# lists and tuples are iterable, ie we can manipulate elements within
L=[1,2,3]
car={"Bob":"Mustang","Judy":"Beetle"}
s="mystring"
f=98.6
b1=True
b2=False

t=(L,s,f,b1,b2,car)
for x in t:
    print(x)
```

#### Ranges

```python
for x in range(5): #prints 0 to 4, cannot take negatives
    print(x)
```

```python
for x in range(1,5): # prints from the 1st up to and not including the last element
    print(x)

for y in range(-5,-1): # prints from the 1st up to and not including the last element
    print(y)
```

```python
for x in range(1,11,2): # from 1 to 9 incrementing by 2
    print(x)

for i in range(10,5,-1):
    print(i)
```

#### Dictionaries (maps)

```python
# A dictionary is a set of'key: value' pairs
# Keys must be hashable (all structures but lists)
car={'George':'Corolla','Karla':'Mustang','Jim':'Accord'}
print(car)
print(type(car))
print(car.keys()) # getting keys
print(car.values()) # getting values
```

```python
car["Harriet"]='Vega' 
print(car) # we can directly add a pair or change values
car["Harriet"]='Corolla'
print(car['Harriet'])
```

```python
car={'George':'Corolla','Karla':'Mustang','Jim':'Accord'}
v=car.pop("George") # popping the key removes the pair and extracts the value
print(car)
print(v)

car={'George':'Corolla','Karla':'Mustang','Jim':'Accord'}
v=car.popitem() # popitem removes AND extracts the pair
print(car)
print(v)
```

```python
car.get("George") # to extract a value from key
```

#### Booleans

```python
t=True
f=False
print(type(t), t) #boolean datatype
print(type(f), f)
b1=not t #not reverses the type
b2=not f
print(b1)
print(b2)
```

```python
# AND operator
b1=t & t
b2=t & f
b3=f & t
b4=f & f
print(b1, b2, b3, b4)
```

```python
# OR operator
b1=t | t
b2=t | f
b3=f | t
b4=f | f
print(b1, b2, b3, b4)
```

```python
# XOR operator (exculsively or)
b1=t ^ t
b2=t ^ f
b3=f ^ t
b4=f ^ f
print(b1, b2, b3, b4)
```

```python
# numbers are compared, and strings are compared by ASCII value https://www.asciitable.com/
x=7
y=8
a='abc'
b='def'
print(x<y,x<=y,x==y,x>=y,x>y)
print(a<b,a<=b,a==b,a>=b,a>b) 
# we can use 'is' in place of ==
print(x is y)
print(a is b)
```
