**Overloading operators**

We can define operators <, <=, ==, +, -, *  etc. for class members.

```python
class thing:
    def __init__(self,m,n,word):
        self.m=m
        self.n=n
        self.word=word
    def __lt__(self,other):
        x=self.m*self.n
        y=other.m*other.n
        return x<y
    def __gt__(self,other):
        x=self.m*self.n
        y=other.m*other.n
        return x>y
    def __le__(self,other):
        x=self.m*self.n
        y=other.m*other.n
        return x<=y
    def __eq__(self,other):
        x=self.word
        y=other.word
        return x==y
    def __add__(self,other):
        m=self.m+other.n
        n=self.n+other.m
        word=other.word+self.word
        return thing(m,n,word)
    def __mul__(self,other):
        m=self.m*other.n
        n=self.n*other.m
        word=self.word+other.word
        return thing(m,n,word)
    def __str__(self):
        s="m = "+str(self.m) + "\n"
        s+="n = "+str(self.n) + "\n"
        s+="word = "+self.word
        return(s)
                
```

```python
thing1=thing(2,13,"duck")
thing2=thing(13,4,"goose")
thing3=thing(5,16,"duck")
print(thing1<thing2)
print(thing1<=thing2)
print(thing1>thing2)
print(thing1==thing2)
print(thing1==thing3)
```

```python
thing4=thing1+thing2
thing5=thing1+thing3
print(thing4)
print(thing5)
```
