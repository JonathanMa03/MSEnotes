```python
import datetime
#
# Transaction data type. This can be a deposit or a withdrawal
#
class Transaction:
    def __init__(self,name,transaction_type, amount, timestamp):
        if transaction_type not in ["Deposit","Withdrawal"]:
            print("Error: transaction type must be Deposit or Withdrawal")
            return
        self.name=name
        self.transaction_type=transaction_type
        self.amount=amount
        self.timestamp=timestamp
    def __str__(self):
        ts=self.timestamp.strftime("%m/%d/%Y %H:%M:%S:%f")
        st=f"{self.name:10s}  {self.transaction_type:10s}  {self.amount:10.2f}  {ts:15s}"
        return(st)
#
# Account data type. This has a name a creation time stamp, a balance and 
# list of transactions.
#
class Account:
    def __init__(self,name,creation_timestamp):
        self.name=name
        self.creation_timestamp=creation_timestamp
        self.balance=0
        self.transaction_list=[] #empty list for all transactions
    def __str__(self): # string representation: used when a print() occurs
        ts=self.creation_timestamp.strftime("%m/%d/%Y %H:%M:%S:%f")
        st=f"{self.name:<10s} {ts:<40s} {self.balance:<10.2f}"
        return(st)
#
# A Bank has accounts and methods for
#    - adding an account
#    - processing a transaction
#    - get all balances for all accounts
#    - get all transactions for an account
#
class Bank:
    def __init__(self):
        self.Accounts=[]
    
    def create_account(self, name, timestamp):
        if name in [x.name for x in self.Accounts]:
            print("Error: this account already exists")
        else:
            # Create new account 
            account=Account(name,timestamp)
            self.Accounts.append(account)
            print(f"Account for {name} created successfully.")
            # print(account)
    def process_transaction(self,transaction):
        name=transaction.name
        #
        # check that this is an account holder
        #
        accounts_with_name=[x for x in self.Accounts if x.name==name]
        if len(accounts_with_name)==0:
            print(f"{name} does not have an account")
            return
        account=accounts_with_name[0]
        
        if transaction.transaction_type=="Deposit":
            if transaction.amount > 0:
                account.balance += transaction.amount
                account.transaction_list.append(transaction)
                print(f"   ${transaction.amount} deposited successfully. New balance: ${account.balance:.2f}")
            else:
                print("   Deposit rejected amount must be positive.")
        elif transaction.transaction_type=="Withdrawal":
            if 0 < transaction.amount <= account.balance:
                account.balance -= transaction.amount
                account.transaction_list.append(transaction)
                print(f"   ${transaction.amount} withdrawn successfully. New balance: ${account.balance:.2f}")
            else:
                print("   Withdrawal rejected: isufficient funds or invalid amount")
        else:
            print("Transaction must be Deposit or Withdrawal")
            return
    def get_all_balances(self):
        print(f"{"Name":<10s} {"Created":<40s} {"Balance":<10s}")
        for account in self.Accounts:
            print(account)

    def show_transactions(self,name):
        accounts_with_name=[x for x in self.Accounts if x.name==name]
        if len(accounts_with_name)==0:
            print(f"{name} does not have an account")
            return
        account=accounts_with_name[0]
        for t in account.transaction_list:
            print(t)
```

```python
# create a gauss fuinction
import numpy as np
class normal_pdf: # define the class and attributes
    def __init__(self,mu, sigma):
        self.mu=mu
        self.sigma=sigma
# use the __call__ to actually return something (Call the function)
    def __call__(self,x):
        return (1/(self.sigma*np.sqrt(2*np.pi)))*np.exp(-.5*((x-self.mu)/self.sigma)**2)
f=normal_pdf(0,1) # set parameters
f(0.68) # call function
```

```python
# Length method
# same as len(x) bc python interprets len(x) as x.__len__()
class Bag:
    def __init__(self, items):
        self.items = items
    def __len__(self):
        return len(self.items)

b = Bag([1, 2, 3])
print(len(b))  # 3
```

```python
# Stringify method
# Defines the human-readable string representation of an object, used by print() and str().
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
    def __str__(self):
        return f"'{self.title}' by {self.author}"

b = Book("1984", "George Orwell")
print(b)  # '1984' by George Orwell
```

```python
# Call Method
#If a class defines __call__, you can use its instances with parentheses () as if they were functions.
class Multiplier:
    def __init__(self, factor):
        self.factor = factor
    def __call__(self, x):
        return x * self.factor

m = Multiplier(3)
print(m(10))  # 30  (same as m.__call__(10))
```

```python
import datetime
class trade:
    """
    A trade object stores information about a trade.
    ntrades keeps track of the number of times a trade is instantiated.
    """
    ntrades=0
    nstr=0
    TSlist=[]
    def __init__(self,timestamp, symbol,buyerid,sellerid,price,quantity):
        self.timestamp=timestamp
        trade.TSlist.append(timestamp)
        self.symbol=symbol
        self.buyerid=buyerid
        self.sellerid=sellerid
        self.price=price
        self.quantity=quantity
        trade.ntrades=trade.ntrades+1
    def __str__(self):
        print("in __str__ \n")
        timestr=self.timestamp.strftime("%m/%d/%Y %H:%M:%S:%f")[:-3]
        s="TimeStamp: " + timestr + "\n"
        s+="Symbol: " + self.symbol + "\n"
        s+="BuyerId: = " + self.buyerid + "\n"
        s+="SellerId: " + self.sellerid + "\n"
        s+="Price: " + str(self.price) + "\n"
        s+="Quantity " + str(self.quantity) + "\n"
        trade.nstr+=1
        return(s)
```

#### Derived Classes

```python
# a derived class can override any method in the super (base) class except for the constructor (__init__)
# --- Base class definition ---
class Animal:
    def __init__(self, name):
        self.name = name             # Attribute common to all animals
    
    def speak(self):
        # A generic method — can be overridden by derived classes
        print(f"{self.name} makes a sound.")
    
    def info(self):
        print(f"This is an animal named {self.name}.")


# --- Derived class: Dog ---
class Dog(Animal): #DerivedClass(SuperClass)
    def __init__(self, name, breed):
        # Call the parent class constructor using super()
        super().__init__(name)
        self.breed = breed           # Extra attribute specific to Dog
    
    # Override the speak() method
    def speak(self):
        print(f"{self.name} says: Woof!")
    
    def info(self):
        # Extend parent's info() method
        super().info()               # Call the base version first
        print(f"{self.name} is a {self.breed} dog.")


# --- Derived class: Cat ---
class Cat(Animal):
    def __init__(self, name, color):
        super().__init__(name)
        self.color = color
    
    def speak(self):
        print(f"{self.name} says: Meow!")
    
    def info(self):
        super().info()
        print(f"{self.name} is a {self.color} cat.")
```

```python
# --- Example usage ---
# Create one Dog and one Cat object
dog1 = Dog("Buddy", "Golden Retriever")
cat1 = Cat("Misty", "gray")

# Call methods on both
dog1.info()     # Demonstrates extended method from Dog
dog1.speak()    # Overridden method

print()

cat1.info()     # Demonstrates extended method from Cat
cat1.speak()    # Overridden method

print()
```

```python
# --- Demonstrate polymorphism ---
animals = [dog1, cat1]
for a in animals:
    a.speak()   # Each calls its own version of speak()
```
