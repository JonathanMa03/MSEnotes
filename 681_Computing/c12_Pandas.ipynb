# **Pandas Series**

The pandas module has many features that are useful in data science applications. Perhaps the most important object is a data frame, especially the 2d (spreadsheet-like) data frames.

Since pandas data frames use pandas series as column, it is a good idea to first understand how to work with pandas series.

**Constructing pandas series**

We can construct a pandas series by providing a list of numbers, just like we do for a numpy array.

```python
import pandas as pd

x=pd.Series([1.4,2.5,6.9,7.4,1.2,9.1,3.5])
u=pd.Series([2,5,7,3])
v=pd.Series([True,False,True])
print(x)
print(u)
print(v)
```

```python
import numpy as np
v=np.array([2.5,6.7,8.9,2.3])
w=pd.Series(v)
print(w)

v=np.array([2,5,6,7],dtype=np.int64)
w=pd.Series(v)
print(w)
```

Note that when we print the series, we get an _index_ column and we get a data type. 

We'll have more to see about indices soon.

Pandas series have much of the functionality of numpy arrays.

```python
import numpy as np
import pandas as pd
x=pd.Series([1.4,2.5,6.9,7.4,1.2,9.1,3.5])
y=pd.Series([3.2,7.5,6.1,7.9,1.4,9.9,-1.5])
z=2.*x+3.*y+10
w=x/y
h=np.sin(y)
print(y)
print(z)
print(w)
print(h)
```

**Referring to elements by position**

We can refer to elements of a pandas series by _position_, using the .iloc method. The arguments to iloc can be an integer, a slice object, a boolean array, or a list of positions.

```python
import pandas as pd
import numpy as np
x=pd.Series(np.random.uniform(0,1,10))
print(x)
print("\n")

print(x.iloc[3])
print("\n")

print(x.iloc[2:5])
print("\n")
y=x.iloc[2:5]
print(type(y))

print(x.iloc[[2,4,7,2]])
print("\n")
```

When conversion using a provided converter is not available, we can write our own converter and use the apply method.

```python
import datetime
def my_str_to_date(dstr):
    return(datetime.datetime.strptime(dstr,"%Y_%m_%d").date())
my_str_to_date("1955_04_12")
```

```python
dates=pd.Series(['1955_04_12','1921_09_06','1925_08_13','1958_12_15'])
dates2=dates.apply(my_str_to_date)
print(dates2)
print(type(dates2[0]))
```

**Rolling**

Rolling means create an object by grouping windows of data in a series. We typically follow up by calculating something for each group of values.

For example, a moving average with a window size of 2 is obtained by taking two successive values at a time and averaging them.

```python
import pandas as pd
import numpy as np
x=pd.Series([0.,2.,4.,5.,3.,1.,0.])
x=x.cumsum()
print(x)
y=x.rolling(window=3).quantile(.75)
print(y)
```

**This is a critical tool in smoothing**

# **Pandas Data Frames**

A pandas data frame is a 2-dimensional array in which the columns have common indices. We usually use such an object when we have a collection of observations/items and a fixed collection of variables/attributes that describe each of the observations.

Consider a convenience store and data for each possible transaction. A transaction has a transaction ID, a customer ID, the amount of money associated with the sale, the amount of tax associated with the sale, and a date time stamp.

The data can be stored as 3 separate dictionaries with the transcation ID as the key. A key point here is that the key is common to all of the variables.

```python
customer={'T12341':"George",'T12342':"Thomas",'T12343':"Josephine",'T12344':"Steve",'T12345':"James"}
amount={'T12341':12.75,'T12342':10.89,'T12343':4.67,'T12344':11.21,'T12345':4.00}
tax={'T12341':.64,'T12342':.55,'T12343':.24,'T12344':.46,'T12345':.15}
newcol={'T12341':[.64,.35],'T12342':55,'T12343':.24,'T12344':"dwjdkw",'T12345':{"g":.15,"h":20}}
datetime={'T12341':'12:23:2018:12:38','T12342':'12:23:2018:12:45','T12343':'12:23:2018:12:45','T12344':'12:23:2018:12:47','T12345':'12:23:2018:12:49'}
```

```python
import pandas as pd
df=pd.DataFrame({'customer':customer,'amount':amount,'tax':tax, 'datetime':datetime}) # "newcol":newcol})
print(df)
```

We can use an existing column to create a new index. We can do this "inplace" meaning that we modify the existing data frame, or we can create a new data frame object.

```python
import numpy as np
df=pd.DataFrame(np.random.normal(10, 5, (10,3)),
            columns=['x', 'y','z'])
df
```

```python
import numpy as np
df=pd.DataFrame(np.random.normal(10, 5, (10,3)),
             columns=['x', 'y','z'])
df2=df.set_index('x')
df2
```

```python
import numpy as np
df=pd.DataFrame(np.random.normal(10, 5, (10,3)),
             columns=['x', 'y','z'])
df.set_index('x',inplace=True)
df.head(10)
```

**Copying**

As usual, remember you might want to copy!

```python
import numpy as np
df=pd.DataFrame(np.random.normal(10, 5, (10,3)),
             columns=['x', 'y','z'])
df['myindex']=['a','b','c','d','e','f','g','h','i','j']
df2=df # df2 is a reference to the same object as df
df2.set_index('myindex',inplace=True)
df.head()
print(df)
```

```python
import numpy as np
df=pd.DataFrame(np.random.normal(10, 5, (10,3)),
             columns=['x', 'y','z'])
df['myindex']=['a','b','c','d','e','f','g','h','i','j']
df2=df.copy()
df2.set_index('myindex',inplace=True)
df.head()
print(df)
```

A good index usually has the property that it identifies a unique observation. So when setting the index, we might wish to perform an integrity check.

```python
import numpy as np
df=pd.DataFrame(np.random.normal(10, 5, (10,3)),
             columns=['x', 'y','z'])
df['myindex']=['a','b','c','d','e','f','g','h','a','j']

try:
    df.set_index('myindex',inplace=True,verify_integrity=True)
except ValueError:
    print("integrity check failed")
df
```

That said, uniqueness is not a strict requirement.

```python
import numpy as np
df=pd.DataFrame(np.random.normal(10, 5, (10,3)),
             columns=['x', 'y','z'])
df['myindex']=['a','b','c','d','e','f','g','h','a','j']
try:
    df.set_index('myindex',inplace=True,verify_integrity=False)
except ValueError:
    print("integrity check failed")
df
```

We can revert back to the default index (0,1,...) using the reset_index method.

When we do this, the old index is retained as a column.

```python
import numpy as np
import pandas as pd
df=pd.DataFrame(np.random.normal(10, 5, (10,3)),
             columns=['x', 'y','z'],
               index=['a','b','c','d','e','f','g','h','i','j'])
print(df)
print(df.columns)

df.reset_index(inplace=True)
print(df)
print(df.columns)
```

We can _reindex_ a data frame by supplying a list of values.

In the following example, we sort the index and use that as the new index.

Key point: this does not change which index value is associated with which row of values in the data frame.

```python
import numpy as np
import pandas as pd
df=pd.DataFrame(np.random.normal(10, 5, (10,3)),
             columns=['x', 'y','z'],
             index=['g','c','h','b','a','e','f','i','d','j'])
print(df)
#L=list(df.index) # list of index values
#print(L)
#print(sorted(L))
df=df.reindex(sorted(df.index))
print(df)
```

The indices do not have to all be included.

```python
import numpy as np
df=pd.DataFrame(np.random.normal(10, 5, (10,3)),
             columns=['x', 'y','z'],
             index=['g','c','h','b','a','e','f','i','d','j'])
print(df)
L=['g','a','e']
print(L)
df=df.reindex(L)
print(df)
```

Repeats are not allowed.

```python
import numpy as np
df=pd.DataFrame(np.random.normal(10, 5, (10,3)),
             columns=['x', 'y','z'],
             index=['g','c','h','b','a','e','f','i','d','j'])
print(df)
L=['a','g','a','g','b','b']
print(L)
df=df.reindex(L)
print(df)
```
