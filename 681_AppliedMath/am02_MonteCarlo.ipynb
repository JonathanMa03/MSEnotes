# The Delta Method — Notes

## **1. Overview**

The **Delta Method** is a fundamental tool for obtaining **approximations to moments** (mean, variance, covariance) of random variables that are smooth functions of other random variables with known moments.

The key idea:
If a random variable $X$ is concentrated near its mean $\mu_X$, and $f$ is a smooth function, then the distribution of $f(X)$ is approximately concentrated near $f(\mu_X)$.  
This allows us to approximate:
- $E[f(X)]$
- $\operatorname{Var}[f(X)]$
- and in higher dimensions, $\operatorname{Cov}[f(U,V)]$

---

## **2. Delta Method for the Mean (1-D Case)**

Suppose $X$ has mean $\mu_X$ and variance $\sigma_X^2$.  
We assume $\sigma_X$ is small, so $X$ is concentrated around $\mu_X$.  
Let $f$ be a sufficiently smooth function near $\mu_X$.

We use a **second-order Taylor expansion** about $\mu_X$:

$$
f(x) \approx f(\mu_X) + f'(\mu_X)(x - \mu_X) + \tfrac{1}{2}f''(\mu_X)(x - \mu_X)^2
$$

Substitute $X$ for $x$:

$$
f(X) \approx f(\mu_X) + f'(\mu_X)(X - \mu_X) + \tfrac{1}{2}f''(\mu_X)(X - \mu_X)^2
$$

Take expectations:

$$
E[f(X)] \approx f(\mu_X) + \tfrac{1}{2} f''(\mu_X) E[(X - \mu_X)^2]
$$

Since $E[(X - \mu_X)^2] = \sigma_X^2$, we get the **approximation for the mean**:

$$
E[f(X)] \approx f(\mu_X) + \tfrac{1}{2}f''(\mu_X)\sigma_X^2
$$

Intuition:
- If $X$ is tightly concentrated near $\mu_X$, then $E[f(X)] \approx f(\mu_X)$.
- The correction term accounts for the curvature (nonlinearity) of $f$.

---

### **Example 1: Area of a Circle**

Let $R$ be the random radius of a circle with mean $\mu_R$ and variance $\sigma_R^2$.  
We want the mean of the area $A = \pi R^2$.

Here $f(r) = \pi r^2$, so:
- $f'(r) = 2\pi r$
- $f''(r) = 2\pi$

Using the delta method:

$$
E[\pi R^2] \approx \pi \mu_R^2 + \pi \sigma_R^2
$$

This is exact, since $E[R^2] = \operatorname{Var}(R) + (E[R])^2$.

---

### **Example 2: Volume from Measured Weight**

Let the measured **weight** $W$ have mean $\mu_W$ and variance $\sigma_W^2$, and let the **density** be $d$.  
The estimated **volume** is:

$$
V = f(W) = \frac{1}{d} W^{1/3}
$$

Compute derivatives:
- $f'(w) = \frac{1}{3d} w^{-2/3}$
- $f''(w) = -\frac{2}{9d} w^{-5/3}$

Using the delta method:

$$
E[V] \approx \frac{1}{d} \mu_W^{1/3} - \frac{1}{9d}\mu_W^{-5/3}\sigma_W^2
$$

So the curvature (negative second derivative) decreases the expected value.

---

## **3. Delta Method for the Variance (1-D Case)**

To approximate the **variance of $f(X)$**, we use only the **first-order** Taylor expansion:

$$
f(x) \approx f(\mu_X) + f'(\mu_X)(x - \mu_X)
$$

Substitute $X$ and take variances:

$$
\operatorname{Var}(f(X)) \approx [f'(\mu_X)]^2 \operatorname{Var}(X) = [f'(\mu_X)]^2 \sigma_X^2
$$

---

### **Example 1 (revisited): Circle Area**

$f(r) = \pi r^2$, $f'(r) = 2\pi r$

$$
\operatorname{Var}(\pi R^2) \approx (2\pi\mu_R)^2 \sigma_R^2 = 4\pi^2 \mu_R^2 \sigma_R^2
$$

---

### **Example 2 (revisited): Volume Estimate**

$f(w) = \frac{1}{d}w^{1/3}$, $f'(w) = \frac{1}{3d} w^{-2/3}$

$$
\operatorname{Var}\left(\frac{1}{d}W^{1/3}\right) \approx \frac{1}{9d^2}\mu_W^{-4/3}\sigma_W^2
$$

---

## **4. The 2-D Case (Functions of Two Variables)**

Let $U$ and $V$ be random variables with means $\mu_U, \mu_V$, variances $\sigma_U^2, \sigma_V^2$, and covariance $\sigma_{U,V}$.  
Define $W = f(U,V)$ for a smooth function $f$.

We use a **second-order Taylor expansion** around $(\mu_U, \mu_V)$:

$$
\begin{aligned}
f(u,v) &\approx f(\mu_U,\mu_V)
+ \frac{\partial f}{\partial u}(\mu_U,\mu_V)(u - \mu_U)
+ \frac{\partial f}{\partial v}(\mu_U,\mu_V)(v - \mu_V) \\
&\quad + \frac{1}{2}\frac{\partial^2 f}{\partial u^2}(\mu_U,\mu_V)(u - \mu_U)^2
+ \frac{1}{2}\frac{\partial^2 f}{\partial v^2}(\mu_U,\mu_V)(v - \mu_V)^2
+ \frac{\partial^2 f}{\partial u\partial v}(\mu_U,\mu_V)(u - \mu_U)(v - \mu_V)
\end{aligned}
$$

---

### **Approximation for the Mean**

Substitute $U, V$ and take expectations:

$$
E[f(U,V)] \approx f(\mu_U,\mu_V)
+ \frac{1}{2}\frac{\partial^2 f}{\partial u^2}\sigma_U^2
+ \frac{1}{2}\frac{\partial^2 f}{\partial v^2}\sigma_V^2
+ \frac{\partial^2 f}{\partial u\partial v}\sigma_{U,V}
$$

All partial derivatives are evaluated at $(\mu_U,\mu_V)$.

---

### **Approximation for the Variance**

Using the **first-order** expansion:

$$
f(U,V) \approx f(\mu_U,\mu_V)
+ \frac{\partial f}{\partial u}(\mu_U,\mu_V)(U - \mu_U)
+ \frac{\partial f}{\partial v}(\mu_U,\mu_V)(V - \mu_V)
$$

Then:

$$
\operatorname{Var}(f(U,V)) \approx
\left(\frac{\partial f}{\partial u}\right)^2\sigma_U^2
+ \left(\frac{\partial f}{\partial v}\right)^2\sigma_V^2
+ 2\frac{\partial f}{\partial u}\frac{\partial f}{\partial v}\sigma_{U,V}
$$

This form explicitly depends on the **covariance** between $U$ and $V$.

---

## **5. Example — Ratios**

Let $f(u,v) = \dfrac{v}{u}$.

Partial derivatives:

$$
\frac{\partial f}{\partial u} = -\frac{v}{u^2}, \qquad
\frac{\partial f}{\partial v} = \frac{1}{u}
$$

Second derivatives:

$$
\frac{\partial^2 f}{\partial u^2} = \frac{2v}{u^3}, \quad
\frac{\partial^2 f}{\partial v^2} = 0, \quad
\frac{\partial^2 f}{\partial u\partial v} = -\frac{1}{u^2}
$$

Plug into the delta method formulas:

### Approximate Mean

$$
E\!\left[\frac{V}{U}\right]
\approx \frac{\mu_V}{\mu_U}
+ \frac{\mu_V}{\mu_U^3}\sigma_U^2
- \frac{1}{\mu_U^2}\sigma_{U,V}
$$

### Approximate Variance

$$
\operatorname{Var}\!\left(\frac{V}{U}\right)
\approx \frac{\mu_V^2}{\mu_U^4}\sigma_U^2
+ \frac{1}{\mu_U^2}\sigma_V^2
- \frac{2\mu_V}{\mu_U^3}\sigma_{U,V}
$$

## **6. Functions of Sample Means**

Let $X_1, \ldots, X_n$ be iid with mean $\mu_X$ and variance $\sigma_X^2 < \infty$.  
The sample mean $\bar{X}$ satisfies:

$$
P(|\bar{X} - \mu_X| > \varepsilon) \le \frac{\sigma_X^2}{n\varepsilon^2} \to 0 \text{ as } n \to \infty
$$

Thus $\bar{X}$ becomes concentrated near $\mu_X$.  
If $f:\mathbb{R} \to \mathbb{R}$ is smooth, the delta method provides approximations to $E[f(\bar{X})]$ and $\operatorname{Var}[f(\bar{X})]$.

It can also be applied to **functions of two sample means**, e.g. ratios.

---

### **Example: Ratio of Two Sample Means**

Let $\bar{Y}$ and $\bar{X}$ be sample means from independent samples:
- $E[\bar{Y}] = \mu_Y$, $E[\bar{X}] = \mu_X$
- $\operatorname{Var}(\bar{Y}) = \sigma_Y^2 / n$
- $\operatorname{Var}(\bar{X}) = \sigma_X^2 / n$
- $\operatorname{Cov}(\bar{X}, \bar{Y}) = 0$

Estimator for $\mu_Y / \mu_X$ is $\bar{Y}/\bar{X}$.

Using the delta method for ratios:

$$
E\!\left[\frac{\bar{Y}}{\bar{X}}\right] \approx
\frac{\mu_Y}{\mu_X} + \frac{\mu_Y\sigma_X^2}{\mu_X^3 n}
$$

$$
\operatorname{Var}\!\left(\frac{\bar{Y}}{\bar{X}}\right) \approx
\frac{\mu_Y^2\sigma_X^2}{\mu_X^4 n} + \frac{\sigma_Y^2}{\mu_X^2 n}
$$

---

## **7. Notes**

- The Delta Method extends easily to **multivariate** random variables using vector and matrix calculus.  
- It is commonly applied in:
  - Nonlinear functions of estimators
  - Transformations of sample means
  - Logarithms, ratios, or powers of random variables
- The approximations improve as the underlying distributions become more concentrated (i.e., smaller variances or larger sample sizes).

## **1. Overview**

The **Monte Carlo Method** is an approach to solving problems that involve **stochastic models** — systems with inherent randomness.  
We have a random object $X$ that takes values in some set or space $\mathcal{X}$.

Typical choices for $\mathcal{X}$:

- A finite set  
- The real line $\mathbb{R}$  
- A finite-dimensional Euclidean space $\mathbb{R}^k$  
- A function space (e.g., continuous paths or functions)

---

## **2. Examples**

### Coin Tossing

- A single toss: $X \in \{0,1\}$
- $p = P(X = 1)$
- Repeated tosses: $X = (X_1,\ldots,X_n) \in \{0,1\}^n$

Typical questions:
- Probability of seeing a run of 15 consecutive heads in 100 flips?
- Average length of the longest run?

---

### Shuffled Deck of Cards

- $X = (c_{\pi_1}, c_{\pi_2}, \ldots, c_{\pi_{52}})$
- $(\pi_1,\ldots,\pi_{52})$ is a random permutation of $1,\ldots,52$
- $X \in S_{52}$: the space of all permutations

Typical questions:
- Probability that all four aces end up in one bridge hand?  
- Expected number of suits represented in one hand?

---

### Stochastic Differential Equation (SDE) Solution

- $X = \{X_t, t \in [0,T]\}$: a random function  
- $X \in C[0,T]$: continuous functions on $[0,T]$

Applications:

- Portfolio values over time  
- Temperature or chemical concentration levels  

Typical questions:
- What is $P(X_t > \tau_U \text{ and } X_t < \tau_L)$?
- Compute $E\!\left[\int_0^T X_t^2\,dt\right]$.

---

## **3. The General Problem**

We want to compute an expectation

$$
\mu = E[g(X)]
$$

for some function $g:\mathcal{X}\to\mathbb{R}$.

If $g(x)$ is an **indicator function** $I_A(x)$, then

$$
E[g(X)] = P[X \in A]
$$

So estimating a probability is a special case of estimating an expectation.

---

## **4. Monte Carlo Estimation**

Simulate $n$ iid samples $X^{(1)},\ldots,X^{(n)}$ of $X$.

Estimator:

$$
\hat{\mu}_n = \frac{1}{n}\sum_{i=1}^n g(X^{(i)})
$$

**Strong Law of Large Numbers:**

$$
\hat{\mu}_n \to \mu \quad \text{with probability 1 as } n\to\infty
$$

**Central Limit Theorem:**

If $\sigma^2 = \operatorname{Var}[g(X)] < \infty$,

$$
\sqrt{n}(\hat{\mu}_n - \mu) \xrightarrow{d} N(0,\sigma^2)
$$

Approximate confidence interval (for large $n$):

$$
\mu \in \hat{\mu}_n \pm z_{\alpha/2}\frac{\sigma}{\sqrt{n}}
$$

where $z_{\alpha/2} = \Phi^{-1}(1-\alpha/2)$ is the standard-normal critical value.

---

## **5. Random Number Generation**

Monte Carlo methods depend on generating **independent random samples**.

Most environments provide a function returning pseudo-random **Uniform(0,1)** values.

These uniform variables are the basic building blocks for constructing more complex random variables.

---

## **6. Generating Random Variables**

### **A. Inversion Method**

If the cumulative distribution function $F_X$ is strictly increasing and continuous, then

$$
X = F_X^{-1}(U), \qquad U\sim \mathrm{Uniform}(0,1)
$$

gives $X \sim F_X$.

**Proof**

$$
P(X \le x) = P(F_X^{-1}(U) \le x) = P(U \le F_X(x)) = F_X(x)
$$

#### Example — Exponential Distribution

For $X \sim \mathrm{Exponential}(\lambda)$:

$$
F_X(x) = 1 - e^{-\lambda x}, \quad x>0
$$

$$
F_X^{-1}(u) = -\frac{\ln(1-u)}{\lambda}
$$

Hence $X = -\frac{\ln(1-U)}{\lambda} \sim \mathrm{Exponential}(\lambda)$.

---

### **B. Acceptance–Rejection Method**

Used when inversion is impractical.

Goal: generate $X \sim f(x)$ given a simpler density $g(x)$ such that

$$
f(x) \le c\,g(x) \quad \forall x
$$

for some constant $c$.

**Algorithm**
Repeat:
Generate X ~ g
Generate U ~ Uniform(0,1)
Until c * U * g(X) <= f(X)
Return X
Expected number of iterations = $c$.  
Choose $g$ and $c$ so $c$ is as small as possible.

---

#### Example — Standard Normal Distribution

Target density:

$$
f(x) = \frac{1}{\sqrt{2\pi}}e^{-x^2/2}
$$

Proposal density:

$$
g(x) = \tfrac12 e^{-|x|} \quad \text{(Laplace distribution)}
$$

**Sampling $g$:**

1. Generate $Y \sim \mathrm{Exponential}(1)$ (via inversion)  
2. Generate $V \sim \mathrm{Uniform}(0,1)$  
3. Set  
   $$
   X = 
   \begin{cases}
   Y, & V \le 1/2\\[3pt]
   -Y, & V > 1/2
   \end{cases}
   \Rightarrow X \sim g
   $$

Find $c$ so that $ f(x) \le c\,g(x) $:

$$
c \ge \frac{f(x)}{g(x)} = \frac{2}{\sqrt{2\pi}}e^{-|x| + x^2/2}
$$

By symmetry, consider $x>0$:

$$
c = \max_x \frac{2}{\sqrt{2\pi}}e^{-x^2/2 + x}
     = \frac{\sqrt{2e}}{\sqrt{\pi}} < 1.32
$$

Thus acceptance rate ≈ $1/1.32 \approx 76\%$.

---

## **7. Explanation of the Acceptance–Rejection Method**

The method is justified by three geometric facts:

1. **Uniform sampling under $cg(x)$:**  
   If $g$ is a pdf, sampling $X\sim g$ and, conditional on $X$,  
   $Y\sim \mathrm{Uniform}(0,cg(X))$ gives $(X,Y)$ uniformly distributed  
   in $ D_{cg} = \{(x,y):0\le y \le c g(x)\}$.

2. **Uniform acceptance principle:**  
   If $ D_1 \subset D_2 \subset \mathbb{R}^2 $ have positive area and we sample uniformly  
   from $D_2$ until a point falls in $D_1$, the accepted point is uniform on $D_1$.  
   The number of trials $N$ has geometric distribution  
   $ P(N=k)=p(1-p)^{k-1} $ with $p = |D_1|/|D_2|$;  
   thus $E[N]=1/p=c$.

3. **Marginal distribution property:**  
   If $f$ is a pdf and we sample uniformly from  
   $ \{(x,y):0\le y \le f(x)\} $, then the marginal $X$ has density $f$.

Together these justify that the algorithm yields samples $ X\sim f $.

---

## **8. Summary Table**

| Method | Key Idea | Works Well For | Notes |
|:--------|:-----------|:---------------|:------|
| **Monte Carlo Estimation** | Approximate $E[g(X)]$ by repeated sampling | Any integrable random variable | Accuracy ∝ $1/\sqrt{n}$ |
| **Inversion Method** | Use $X=F^{-1}(U)$ | Distributions with known inverse CDF | Exact but limited |
| **Acceptance–Rejection** | Accept proposed $X\sim g$ with probability $f(X)/(c\,g(X))$ | Complex distributions | Efficiency depends on $c$ |

**Monte Carlo Example**

Assume U and V are independent Uniform(0,1) random variables.
Let 
$$
X = \frac{U^2}{1+V^2}
$$

**Question**

What is the probability that $X>\frac{1}{2}$?

```python
import numpy as np
ntrials=10000

#MC simulates pairs (U,V) and computes X, and checkes how many are greater than 0.5. It uses the frequency as the probability estimate
X=np.zeros(ntrials) # Makes an array with the length=trials

for i in range(ntrials):
    U=np.random.uniform(0,1) # Draws U from (0,1)
    V=np.random.uniform(0,1) #Draws V from (0,1)
    X[i]=(U**2)/(1+V**2) # Computes X

Y=X>.5 #Creates a boolean array, if X>0.5, return TRUE, else FALSE

phat=round(Y.mean(),3) #Converts T/F to 1/0
print("phat = "+str(phat))

sd=Y.std() # Computes SD of bernoulli outcomes
print("sd = "+str(round(sd,3))) 
sd2=np.sqrt(phat*(1-phat)/(ntrials-1)) # SD with a finite sample correction
print("alt version of sd = "+str(round(sd2,3)))

se=sd/np.sqrt(ntrials)
print("se = "+str(se))

cilower=phat-1.96*se
ciupper=phat+1.96*se
print("ci lower = "+str(cilower))
print("ci upper = "+str(ciupper))
```

**Question**

What is the expected value of $e^{-X}$?

```python
import numpy as np
ntrials=10000
X=np.zeros(ntrials)
for i in range(ntrials):
    U=np.random.uniform(0,1)
    V=np.random.uniform(0,1)
    X[i]=(U**2)/(1+V**2)
Y=np.exp(-X) # g(X)'s
muhat=Y.mean()
print("muhat = "+str(muhat))
sd=Y.std()
print("sd = "+str(sd))
se=sd/np.sqrt(ntrials)
print("se = "+str(se))
cilower=muhat-1.96*se
ciupper=muhat+1.96*se
print("ci lower = "+str(cilower))
print("ci upper = "+str(ciupper))
```

```python
import numpy as np
import time
import matplotlib.pyplot as plt

def delta_method(Y, g, g_prime):
    start_time = time.time()

    n = len(Y)
    theta_hat = np.mean(Y)
    var_theta = np.var(Y, ddof=1) / n
    g_hat = g(theta_hat)
    delta_var = (g_prime(theta_hat) ** 2) * var_theta

    elapsed_time = time.time() - start_time
    return g_hat, delta_var, elapsed_time

def monte_carlo(Y, g, n_sim=10000):
    start_time = time.time()

    n = len(Y)
    means = [np.mean(np.random.choice(Y, size=n, replace=True)) for _ in range(n_sim)]
    g_vals = [g(theta) for theta in means]
    g_hat = np.mean(g_vals)
    var = np.var(g_vals, ddof=1)

    elapsed_time = time.time() - start_time
    return g_hat, var, elapsed_time

# g(.) = log(.)
g = np.log
g_prime = lambda x: 1 / x

# Generate data: Y ~ N(10, 1)
np.random.seed(42)
Y = np.random.normal(loc=10, scale=1, size=100)

g_hat_dm, var_dm, time_dm = delta_method(Y, g, g_prime)
g_hat_mc, var_mc, time_mc = monte_carlo(Y, g, n_sim=10000)

# --- Print results ---
print("Delta Method Approximation:")
print(f"  g(x)      = {g_hat_dm:.5f}")
print(f"  Var       ≈ {var_dm:.8f}")
print(f"  Time      = {time_dm:.6f} sec")

print("Monte Carlo Approximation (10,000 bootstraps):")
print(f"  E[g(x)]   ≈ {g_hat_mc:.5f}")
print(f"  Var       ≈ {var_mc:.8f}")
print(f"  Time      = {time_mc:.6f} sec")

# --- Optional Plot ---
import seaborn as sns
sns.histplot([g(np.mean(np.random.choice(Y, size=len(Y), replace=True))) for _ in range(1000)],
             kde=True, bins=50)
plt.axvline(g_hat_dm, color='red', linestyle='--', label='Delta Method Mean')
plt.title("Monte Carlo Sampling Distribution of g(mean(Y))")
plt.xlabel("g(mean(Y))")
plt.legend()
plt.show()
```
